## 1.3 Grab the title of a website


### 1.3.1 First Case

Example Websites: http://example.com/ 

```python
# Call the website through requests library

example_page = requests.get("http://example.com")
```

```python
# Check the type of the varaible

type(example_page)
```

```python
# To see the result/response we got from the url

example_page.text
```

```python
# In order to grab or use this text we need make it as a soup ( with the help of beautifulsoup)
# Soup has ingredients, which we can choose ( ingredients - tags, classes and id's)

soup = bs4.BeautifulSoup(example_page.text,"lxml")
```

```python
# Now we can call the soup

soup
```

```python
# Now to grab an element (title)

soup.select("title") # We got a list beacuse there can be more than one element of same tag
```

```python
soup.select("p")
```

```python
soup.select("h1")
```

```python
# Now to grab the actual text (Convert back to text)

soup.select("title")[0].getText()
```

```python
# Or another way

soup.select("title")[0].text
```

```python
# Let's try to grab the table from the wikipedia website

# Request for the page
wiki_page = requests.get("https://en.wikipedia.org/wiki/Nokia")

# Make a soup out of it
soup = bs4.BeautifulSoup(wiki_page.text,"lxml")

# Select the class we want
table = soup.select(".toctext") # Here we slected the class we want
```

```python
table # All content realted to the class appear here
```

```python
# We got the content in a list format
# We can print all the content from the table with a loop

for tab in table:
    print(tab.getText())
```

```python
# Another Example 
# We want to grab the co-ordinates shown on the wiki page


# Request for the page
wiki_page = requests.get("https://en.wikipedia.org/wiki/Nokia")

# Make a soup out of it
soup = bs4.BeautifulSoup(wiki_page.text,"lxml")

# Select the class we want
coordinate = soup.select(".geo-dec") # Here we slected the class we want

# Show the coordinate

for coor in coordinate:
    print(coor.getText())
```

## 1.5 Grab an image from any website

Here we'll try to pick out the images from a web page.

```python
# We want to grab the nokia headquater image shown on the wiki page


# Request for the page
wiki_page = requests.get("https://en.wikipedia.org/wiki/Nokia")

# Make a soup out of it
soup = bs4.BeautifulSoup(wiki_page.text,"lxml")

# Select the class we want
images = soup.select(".image img") # Here we slected the class we want
```

```python
part = images[1]['src'] # Select the src tag we need
```

```python
part
```

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Midpoint_Nokia_Karaportti.jpg/250px-Midpoint_Nokia_Karaportti.jpg">

```python
# Make request for the specific link
image_link = requests.get(f'https:{part}')
```

```python
# To see the raw content of the file/ image ( binary file)

image_link.content
```

```python
with open('nokia-headquater.jpg', 'wb') as f:
    f.write(image_link.content)
```

```python
# Let's try again with the nokia logo

part = images[0]['src'] # Select the src tag we need

# Make request for the specific link
image_link = requests.get(f'https:{part}')

# Write the binray file
with open('nokia-logo.jpg', 'wb') as f:
    f.write(image_link.content)
```
